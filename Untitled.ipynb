```python
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_pickle('nyc-2017-yellow-taxi-trips-to-airport.pkl.gz')
```

```python
df = df.rename(columns={'tpep_pickup_datetime': 'pickup_datetime','tpep_dropoff_datetime': 'dropoff_datetime'})
```

```python
df = df.rename(columns={'dropff_zone': 'dropoff_zone'})
```

```python
df['trip_duration'] = df.dropoff_datetime - df.pickup_datetime
```

```python
df['dropoff_month'] = df.dropoff_datetime.dt.month
df['dropoff_weekofyear'] = df.dropoff_datetime.dt.weekofyear
df['dropoff_dayofyear'] = df.dropoff_datetime.dt.dayofyear
df['dropoff_dayofmonth'] = df.dropoff_datetime.dt.day
df['dropoff_dayofweek'] = df.dropoff_datetime.dt.dayofweek
df['dropoff_hour'] = df.dropoff_datetime.dt.hour
```

```python
df['pickup_borough_code'] = df.pickup_borough.cat.codes
df['pickup_zone_code'] = df.pickup_borough.cat.codes
df['pickup_service_zone_code'] = df.pickup_borough.cat.codes
df['dropoff_zone_code'] = df.dropoff_zone.cat.codes
df['trip_duration_in_minutes'] = df.trip_duration.dt.total_seconds() / 60
```

```python
dfs = df.sample(frac=0.1, random_state=0)
X = dfs[['dropoff_month', 'dropoff_weekofyear', 'dropoff_dayofyear',
         'dropoff_dayofmonth', 'dropoff_dayofweek', 'dropoff_hour',
         'pickup_borough_code', 'pickup_zone_code', 'pickup_service_zone_code',
         'dropoff_zone_code']]
y = dfs['trip_duration_in_minutes'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
```

```python
%time slr.fit(X_train, y_train)
```

```python
y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)
```

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
plt.scatter(y_train_pred, y_train_pred - y_train, c='steelblue', marker='o', edgecolors='white', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='limegreen', marker='s', edgecolors='white', label='Test data')
plt.xlabel('Predicted trip duration')
plt.ylabel('Residuals')
plt.legend()
```

```python
y_train.shape, y_train_pred.shape,y_test.shape, y_test_pred.shape
```

```python
from sklearn.metrics import mean_squared_error
print("MSE train: {:.3f}, test: {:.3f}".format(mean_squared_error(y_train, y_train_pred),
                                              mean_squared_error(y_test, y_test_pred)))
```

```python
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor(LinearRegression(), max_trials=100, min_samples=10,
                         loss='absolute_loss', residual_threshold=15.0, random_state=0)
```

```python
%time ransac.fit(X, y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
line_X = np.arange(3, 10, 1)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators=100, criterion='mse', random_state=1, n_jobs=-11)
%time forest.fit(X_train, y_train)
```

```python
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)
print("MSE train: {:.3f}, test: {:.3f}".format(mean_squared_error(y_train, y_train_pred),
                                              mean_squared_error(y_test, y_test_pred)))
```

```python
from sklearn.metrics import r2_score
print("R^2 train: {:.3f}, test: {:.3f}".format(r2_score(y_train, y_train_pred),
                                              r2_score(y_test, y_test_pred)))
```
